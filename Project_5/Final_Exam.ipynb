```python
Texas_df = pd.read_csv(os.path.join("./csv","US", "2020-04-11_Texas.csv"))
Cali_df = pd.read_csv(os.path.join("./csv","US", "2020-04-11_California.csv"))
```

```python
merged_left = pd.merge(left=Texas_df, right=Cali_df, left_on='Grocery & pharmacy', right_on='Grocery & pharmacy', how='left')

# The LEFT JOIN produces a complete set of records from DataFrame A (left DataFrame), with the matching records (where available) in DataFrame B (right DataFrame). If there is no match, the left side will contain null. You have to pass left in the how argument of merge() function to do left join:
# here, Texas, California, Humboldt_County(CA) and Sonoma_County (CA) have Grocery & pharmacy value of -14%
# same for Angelina_County(TX), Solano_County(CA), Stanislaus_County(CA), Sutter_County(CA) with value of -4%	
merged_left.head(15)
```
